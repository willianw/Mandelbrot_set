```julia
using Distributed
using PyPlot
```

```julia
const N_HEIGHT = 4000
const N_WIDTH = 1.5 * N_HEIGHT
const X0 = -2
const X1 = 1
const Y0 = -1
const Y1 = 1
```

```julia
@everywhere steps_dist = []
```

```julia
@everywhere function converge(x)
	z = 0
	i = 0
	while i < 20 && abs(z) < 2
		z = z ^ 2 + x
		i += 1
	end
    push!(steps_dist, i)
	return i
end
```

```julia
function coordinates(i, j)
	( (X0 + i * (X1 - X0) / N_WIDTH)
	+ (Y0 + j * (Y1 - Y0) / N_HEIGHT) * 1im)
end
```

```julia
M = [ coordinates(j, i) for i = 1:N_HEIGHT, j = 1:N_WIDTH]

img = pmap(converge, M);
```

```julia
figure(figsize=(20, 8))
imshow(img, cmap="gist_stern", aspect="auto")
colorbar()
show()
```

```julia
figure(figsize=(20, 4), dpi=80)
hist(steps_dist, bins=maximum(steps_dist))
show()
```
